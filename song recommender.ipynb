```python
import csv
```

```python
data=[]
with open('song_data.csv',mode='r',encoding='utf-8')as file:
    reader=csv.DictReader(file)
    for row in reader:        
        song_name=row["song_name"],
        song_popularity=row["song_popularity"]
        data.append(
            {"song_name": song_name, "song_popularity": song_popularity}
        )
       
data[:10]
    
```
